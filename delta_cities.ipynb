```python
import pandas as pd
```

```python
url = "https://en.wikipedia.org/wiki/List_of_Delta_Air_Lines_destinations"
tables = pd.read_html(url)
# Drop the irrelevant "Refs" column.
destinations = tables[0].drop("Refs", axis=1)
destinations.head().style.hide()
```

```python
us_airports = destinations[destinations["Country / Territory"].str.contains("United States", regex=False)].style.hide()
us_airports
```
